# TASK-2 
(The cost column in dataframe "Project" has some missing values. Your task is to compute these missing values. Replace the missing values by running average. You should use "For" loop for this task.)

```python
df_1 = pd.read_csv(r'C:\Users\91960\OneDrive\Desktop\roadmaps\Python\Capstone\Initial tables\Project_Dataframe.csv') #reading the csv files and storing in df_1 variable

running_avg = 0 #Initialize a variable running_avg to store the running average of the 'Cost' column
count = 0 #Initialize a variable count to keep track of the number of non-missing values encountered

# Iterate over each row in the DataFrame 'df_1' along with their corresponding index
for index, row in df_1.iterrows():
    # Check if the value in the 'Cost' column is missing (NaN)
    if pd.isnull(row['Cost']):
        # If the value is missing, replace it with the current running average
        df_1.at[index, 'Cost'] = running_avg
    else:
        # If the value is not missing, update the running average based on the current value
        # The running average is calculated using the formula: (previous_avg * count + current_value) / (count + 1)
        running_avg = (running_avg * count + row['Cost']) / (count + 1)
        # Increment the count of non-missing values encountered
        count += 1

df_1['Cost'] = pd.to_numeric(df_1['Cost'], errors='coerce')     

pd.options.display.float_format = '{:.0f}'.format
    
    
df_1.to_csv(r'C:\Users\91960\OneDrive\Desktop\roadmaps\Python\Capstone\Updated tables\df_1.csv',index=False)

df_1
```

# TASK-4 
(Join all three dataframes in one single dataframe.Name it "Final")

```python
# Read the first CSV file into DataFrame "df_1"
df_1 = pd.read_csv(r'C:\Users\91960\OneDrive\Desktop\roadmaps\Python\Capstone\Updated tables\df_1.csv')

# Read the second CSV file into DataFrame "df_2"
df_2 = pd.read_csv(r'C:\Users\91960\OneDrive\Desktop\roadmaps\Python\Capstone\Updated tables\df_2.csv')

# Read the third CSV file into DataFrame "df_3"
df_3 = pd.read_csv(r'C:\Users\91960\OneDrive\Desktop\roadmaps\Python\Capstone\Initial tables\Seniority_Level_Dataframe.csv')

# Merge the DataFrames "df_1", "df_2", and "df_3" based on the 'ID' column
# First, merge "df_1" and "df_2" on the 'ID' column
# Then, merge the result with "df_3" on the 'ID' column
Final = pd.merge(pd.merge(df_1,df_2,on='ID'),df_3,on='ID')

# Display the updated DataFrame "Final"
Final
```

# TASK-5
(Add a new bonus column in final dataframe. Give 5% bonus concerning project cost only for employees who have finished the project.) 

```python
# Calculate bonus based on project status:
# If the project status is 'Finished' or 'Finish ed', calculate bonus as 5% of project cost; otherwise, set bonus to 0       
Final['Bonus'] = Final.apply(lambda row: row['Cost'] * 0.05 if row['Status'] in ['Finished', 'Finish ed'] else 0, axis=1)

# Display the updated DataFrame "Final"
Final
```

# TASK-6
(Decrease the designation level by 1, whose projects have status "fail". Delete the employee record whose designation level is above 4.)

```python
# Convert 'Status' column to lowercase for easier comparison
Final['Status'] = Final['Status'].str.lower()

# Find the indices where projects have failed
failed_indices = np.where(Final['Status'].str.contains('fail'))[0]

# Decrease the designation level by 1 for those projects
Final.loc[failed_indices, 'Designation Level'] -= 1

# Filter the DataFrame "Final" to keep only the rows where the 'Designation Level' column is less than or equal to 4
Final = Final[Final['Designation Level'] <= 4]

# Display the updated DataFrame "Final"
Final
```

# TASK-7
(Add a new column for gender, than add “Mr.” and “Mrs” to the first name column.)

```python
# Define a list of gender values
gender_values = ['M','M','M','F','F','M','M','M','F','F','F','F','F','F']

# Assign the list of gender values to the 'Gender' column of the DataFrame "Final"
Final['Gender'] = gender_values

# Display the updated DataFrame "Final"
Final
```

```python
# Add a new column 'First Name' to the DataFrame "Final" based on the 'Gender' column
# If 'Gender' is 'M', prepend 'Mr. ' to the 'First Name'; otherwise, prepend 'Mrs. '
Final['First Name'] = Final.apply(lambda row: 'Mr. ' + row['First Name'] if row['Gender'] == 'M' else 'Mrs. ' + row['First Name'], axis=1)

# Display the updated DataFrame "Final"
Final
```

# TASK-8
(Increase designation level by 1 for the employees whose age is more than 29 years using IF condition.)

```python
# Extract the 'Age' column values from the DataFrame and convert it to a NumPy array
age_values = Final['Age'].values

# Find the indices where the age is greater than 29
indices = np.where(age_values > 29)

# Make a copy of the 'Designation Level' column values from the DataFrame and convert it to a NumPy array
designation_level_values = Final['Designation Level'].values.copy()

# Increase the designation level by 1 for employees whose age is greater than 29 using NumPy array indexing
designation_level_values[indices] += 1

# Make a copy of the original DataFrame to avoid modifying the original DataFrame directly
Final = Final.copy()

# Update the 'Designation Level' column in the copied DataFrame with the modified values
Final['Designation Level'] = designation_level_values

# Print the modified DataFrame
Final
```

# TASK-9
(Add the cost of all projects for each Employee and save it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost.)

```python
# Group the DataFrame "Final" by 'ID' and 'First Name', and calculate the sum of the 'Cost' column for each group
TotalProjCost = Final.groupby(['ID','First Name'], as_index=False)['Cost'].sum()

# Rename the 'Cost' column to 'Total cost'
TotalProjCost.rename(columns={'Cost': 'Total cost'}, inplace=True)

# Display the DataFrame "TotalProjCost"
TotalProjCost
```
